```python
import pandas as pd

naruto_filepath = './input/naruto.csv'
naruto_data = pd.read_csv(naruto_filepath)

naruto_data.head()
```

```python
naruto_data.describe()
```

```python
naruto_data.loc[[13, 188, 200], ["Rate", "Title"]]
```

```python
naruto_data.loc[naruto_data["Rate"] > 8.0, ["Rate", "Title"]]
```

## Best episodes:

```python
best_episode = naruto_data["Rate"].max()
best_episode_row = naruto_data.loc[naruto_data["Rate"] == best_episode]
best_episode_row
```

## Worst episodes:

```python
worst_episode = naruto_data["Rate"].min()
worst_episode_row = naruto_data.loc[naruto_data["Rate"] == worst_episode]
worst_episode_row
```
