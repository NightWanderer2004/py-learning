```python
import pandas as pd

customers_filepath = './input/customers.csv'
customers_data = pd.read_csv(customers_filepath, index_col='CustomerID')

customers_data.head()
```

```python
customers_data.describe()
```

### Max age:

```python
max_age = customers_data['Age'].max()
max_age_row = customers_data[customers_data['Age'] == max_age]
max_age_row
```

## Min age:

```python
min_age = customers_data['Age'].min()
min_age_row = customers_data[customers_data['Age'] == min_age]
min_age_row
```

## Type of customer:

```python
mean_value = customers_data['Spending Score (1-100)'].mean()

def check_type(spending_score):
    return 'Low' if spending_score < mean_value else 'High'


customers_data['Spending rate'] = customers_data['Spending Score (1-100)'].apply(check_type)
customers_data.head()
```

```python
high_spending_customers = customers_data[customers_data['Spending rate'] == 'High']
low_spending_customers = customers_data[customers_data['Spending rate'] == 'Low']

print(f"{len(high_spending_customers)} customers spend more than the average.")
print(f"{len(low_spending_customers)} customers spend less than the average.")
```
