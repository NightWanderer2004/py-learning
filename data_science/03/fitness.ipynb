```python
import pandas as pd
import matplotlib.pyplot as plt
fitness_data = pd.read_csv('./input/fitness_class_2212.csv')
fitness_data.head()
```

### Average num of monthes with us

```python
fitness_data['months_as_member'].mean()
```

### Histogram of clinets weight

```python
plt.figure(figsize=(7, 5))
plt.hist(fitness_data['weight'])
plt.xlabel('Weight (lbs)')
plt.ylabel('Count')
plt.show()
```

### Sum of attended clients

```python
fitness_data.loc[fitness_data["attended"] == 1].shape[0]
# .shape[0] means number of rows
```

### Average weight of the categories

```python
fitness_data.groupby('category')['weight'].mean().reset_index()
```

### Average weight of each day

```python
plt.figure(figsize=(7, 5))
aver_weight = fitness_data.groupby('day_of_week')['weight'].mean()
plt.bar(aver_weight.index, aver_weight.values)
plt.show()
```

### Scatter plot of weight and months as member

```python
colors = {'Cycling': 'tab:orange', 'Yoga': 'g', 'Strength': 'r', 'Aqua': 'b', 'HIIT': 'tab:purple', '-': 'k'}

plt.figure(figsize=(12, 12))
plt.grid()

plt.title("orrelation between weight and time spent as a member")
plt.xlabel("Months as a member")
plt.ylabel("Weight (lbs)")
plt.xlim(0, 70)
plt.ylim(55, 130)

plt.scatter(fitness_data['months_as_member'], fitness_data['weight'], c=fitness_data['category'].map(colors), alpha=0.6)
plt.show()
```
