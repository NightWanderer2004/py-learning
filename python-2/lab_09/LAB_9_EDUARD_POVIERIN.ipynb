### Zadaie 2:

```python
# Interpolacja danych
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/9.0)

# Interpolate the data
f = interp1d(x, y)
x_new = np.linspace(0, 10, num=41, endpoint=True)
y_new = f(x_new)

# Plot the original and interpolated data
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Interpolated Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Transformata Fouriera
from scipy.fft import fft, ifft

import numpy as np
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
y = fft(x)
yinv = ifft(y)

print("Y: ", y)
print("Y inverted: ", yinv)
```

### Zadanie 3:

```python
from sympy import *
x, y, z = symbols('x y z')
init_printing(use_unicode=True)

expr = exp(x*y*z)
deriv = Derivative(expr, x, y, y, z, 4)
deriv
```

### Zadanie 4:
Dla własnych przykładów całek, pochodnych i układów równań zaprezentować wykorzystanie modułu do obliczeń symbolicznych

```python
from sympy import symbols, diff

x = symbols('x')
expr = x**3 + 2*x**2 - 5*x + 1
print("Pochodna: ", expr)
derivative = diff(expr, x)
print("Pochodna wyrażenia:")
derivative
```

```python
integral = integrate(expr, x)
integral
```
